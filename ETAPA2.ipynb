```python
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# 1. PARÁMETROS
# =====================================================

np.random.seed(2026)

N = 100      # estampas distintas

S = 7        # estampas por sobre

R = 10000    # simulaciones

M_valores = [20,25,30,35,40,45,50,60,70,80,90,100,120]

```

```python

# =====================================================
# 2. LISTA PARA PROBABILIDADES
# =====================================================

probabilidades = []
```

```python
# =====================================================
# 3. SIMULACIÓN
# =====================================================

for M in M_valores:

    exitos = 0

    # -------------------------------------------------
    # Repetir 10,000 simulaciones
    # -------------------------------------------------

    for _ in range(R):

        album = set()

        # ---------------------------------------------
        # Comprar exactamente M sobres
        # ---------------------------------------------

        for _ in range(M):

            sobre = np.random.choice(
                np.arange(N),
                size=S,
                replace=False
            )

            # Agregar estampas al álbum
            for estampa in sobre:

                album.add(estampa)

        # ---------------------------------------------
        # Verificar si completó álbum
        # ---------------------------------------------

        if len(album) == N:

            exitos += 1

    # -------------------------------------------------
    # Probabilidad estimada
    # -------------------------------------------------

    prob = exitos / R

    probabilidades.append(prob)

    print(f"M = {M} --> Probabilidad = {prob:.4f}")
```

```python

# =====================================================
# 4. GRÁFICA
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(
    M_valores,
    probabilidades,
    edgecolor='black',
    alpha=0.7
)

# Línea horizontal en 0.5
plt.axhline(
    y=0.5,
    color='red',
    linestyle='--',
    label='Probabilidad 0.5'
)

plt.title("Probabilidad de completar el álbum")

plt.xlabel("Número de sobres (M)")

plt.ylabel("Probabilidad estimada")

plt.legend()

plt.grid(axis='y', alpha=0.3)

plt.show()
```
