```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2026)

N = 100          # estampas distintas

S = 7            # estampas por sobre

PRECIO = 9.50    # precio por sobre

PRESUPUESTO = 1000

R = 10000        # simulaciones
```

```python

exitos = []

sobres_totales = []

distintas_no_exito = []
```

```python
for _ in range(R):

    # -------------------------------------------------
    # Álbum vacío
    # -------------------------------------------------

    album = set()

    gasto = 0

    sobres = 0

    # -------------------------------------------------
    # Comprar sobres
    # -------------------------------------------------

    while (
        gasto + PRECIO <= PRESUPUESTO
        and len(album) < N
    ):

        # Comprar sobre
        sobre = np.random.choice(
            np.arange(N),
            size=S,
            replace=False
        )

        # Agregar estampas
        for estampa in sobre:

            album.add(estampa)

        # Actualizar gasto
        gasto += PRECIO

        # Contar sobres
        sobres += 1

    # -------------------------------------------------
    # ¿Completó álbum?
    # -------------------------------------------------

    if len(album) == N:

        exitos.append(1)

    else:

        exitos.append(0)

        distintas_no_exito.append(len(album))

    # Guardar sobres
    sobres_totales.append(sobres)
```

```python
prob_completar = np.mean(exitos)

esperado_sobres = np.mean(sobres_totales)

esperado_distintas = np.mean(distintas_no_exito)
```

```python
print("RESULTADOS")

print(f"Probabilidad de completar álbum: {prob_completar:.4f}")

print(f"Número esperado de sobres: {esperado_sobres:.4f}")

print(
    f"Número esperado de estampas distintas "
    f"(si no completa): {esperado_distintas:.4f}"
)
```

```python
completo = np.mean(exitos)

no_completo = 1 - completo

plt.figure(figsize=(6,5))

plt.bar(
    ['Completa', 'No completa'],
    [completo, no_completo],
    color=['green', 'red'],
    edgecolor='black'
)

plt.ylabel("Proporción")

plt.title("Probabilidad de completar el álbum")

plt.ylim(0,1)

plt.grid(axis='y', alpha=0.3)

plt.show()
```

# Pregunta 1 
Se necesitan minimo 105 sobres, 1000/9.5. Pero, sin repetidas son 15, porque 1000/7. Este valor es mayor a 15 por lo tanto es suficiente. 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2026)

N = 100          # estampas distintas

S = 7            # estampas por sobre

R = 10000        # simulaciones

SOBRES_CAJA = 104
```

```python
exitos_caja = []
```

```python
for _ in range(R):

    # Álbum vacío
    album = set()

    # -------------------------------------------------
    # Comprar exactamente 104 sobres
    # -------------------------------------------------

    for _ in range(SOBRES_CAJA):

        sobre = np.random.choice(
            np.arange(N),
            size=S,
            replace=False
        )

        # Agregar estampas
        for estampa in sobre:

            album.add(estampa)

    # -------------------------------------------------
    # Verificar si completó álbum
    # -------------------------------------------------

    if len(album) == N:

        exitos_caja.append(1)

    else:

        exitos_caja.append(0)
```

```python
prob_caja = np.mean(exitos_caja)
print("RESULTADOS CAJA")
print(
    f"Probabilidad de completar álbum "
    f"con caja: {prob_caja:.4f}"
)
```

# Pregunta 3 
Sobran Q.25 al comprar la caja, el sobre extra cuesta Q. 9.50 por lo que se pueden comprar 2 sobres más. Se puedehacer una estrategia mixta la cual seria comprar la caja y 2 sorbes extra. Y como se ve en la simulación

```python
import numpy as np
np.random.seed(2026)

N = 100

S = 7

R = 10000

TOTAL_SOBRES = 106

```

```python
exitos_mixto = []
```

```python
for _ in range(R):

    album = set()

    # Comprar 106 sobres
    for _ in range(TOTAL_SOBRES):

        sobre = np.random.choice(
            np.arange(N),
            size=S,
            replace=False
        )

        for estampa in sobre:

            album.add(estampa)

    # Verificar álbum completo
    if len(album) == N:

        exitos_mixto.append(1)

    else:

        exitos_mixto.append(0)
```

```python
prob_mixto = np.mean(exitos_mixto)
print("ESTRATEGIA MIXTA")
print(
    f"Probabilidad de completar álbum "
    f"con 106 sobres: {prob_mixto:.4f}"
)
```
