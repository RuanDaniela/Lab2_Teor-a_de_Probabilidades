```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 980
S = 7
precio_sobre = 9.50
precio_caja = 975
sobres_caja = 104
simulaciones = 10000
```

```python
def abrir_sobre():
    return [random.randint(1, N) for _ in range(S)]
def simular_hasta_porcentaje(porcentaje):
    objetivo = int(N * porcentaje)
    album = set()
    sobres = 0
    while len(album) < objetivo:
        sobre = abrir_sobre()
        album.update(sobre)
        sobres += 1
    return sobres
def simular_album_completo():
    album = set()
    sobres = 0
    while len(album) < N:
        sobre = abrir_sobre()
        album.update(sobre)
        sobres += 1
    return sobres
```

Probabiblidad de faltar mas de 50 estampas despues de 400 sobres

Pregunta 1

```python
contador = 0
```

```python
for i in range(simulaciones):
    album = set()
    for j in range(400):
        sobre = abrir_sobre()
        album.update(sobre)
    faltantes = N - len(album)
    if faltantes > 50:
        contador += 1
probabilidad_faltantes = contador / simulaciones
print('Pregunta 1')
print('Probabilidad de faltar mas de 50 estampas',
      round(probabilidad_faltantes, 4))
```

Cuantos sobres se necesitan para completar el 75%?

Pregunta 2

```python
resultados_75 = []
for i in range(simulaciones):
    sobres = simular_hasta_porcentaje(0.75)
    resultados_75.append(sobres)
media_75 = np.mean(resultados_75)
print('Pregunta 2')
print('Promedio sobres para 75%:', round(media_75, 2))
print('Costo promedio: Q', round(media_75 * precio_sobre, 2))
```

```python
plt.hist(resultados_75, bins = 20)
plt.title('Sobres necesarios para alcanzar 75%')
plt.xlabel('Cantidad de sobres')
plt.ylabel('Frecuencia')
plt.show()
```

Cuantas estampas unicas se obtienen en promedio despues de abrir 100 sobres?

Pregunta 3

```python
estampas_unicas = []
```

```python
for i in range(simulaciones):
    album = set()
    for j in range(100):
        sobre = abrir_sobre()
        album.update(sobre)
    estampas_unicas.append(len(album))
```

```python
media = np.mean(estampas_unicas)
print('Promedio de estampas unicas:',
      round(media, 2))
```
